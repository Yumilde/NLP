### 2. 实验总结

本次实验让我把课堂中的信息抽取流程真正“跑通”了一遍：先做命名实体识别，用 BIO 标注明确实体边界；再在实体基础上抽取主体、谓词、客体关系；最后将结果映射为知识图谱中的节点与边，实现可视化呈现。通过中英文样例测试，我直观看到从非结构化文本到结构化数据的转化过程，也发现了规则法在指代消解、复杂句和嵌套实体上的局限。整体来看，这次实验不仅加深了我对 NER、RE 与图谱构建衔接关系的理解，也提升了我将 NLP 理论落地为可交互系统的实践能力。

##### 2.1 模块 1 观察任务：1. 输入包含人名和机构名的句子（如“Steve Jobs founded Apple”）。观察并对比直观的 UI 高亮与底层的 BIO 标注序列。体会课件中 B（Begin）、I（Inside）、O（Outside）是如何帮助模型确定实体边界的。2. 思考：如果在语料中出现嵌套实体（如“University of California, Los Angeles”，其中“Los Angeles”本身也是地名），简单的单层 BIO 标注会遇到什么挑战？
- 1. 在输入“Steve Jobs founded Apple”后，界面会把“Steve Jobs”标成 Person，把“Apple”标成 Organization。切到 BIO 视图可见：Steve 为 B-PER，Jobs 为 I-PER，founded 为 O，Apple 为 B-ORG。这种“起点+延续+非实体”的标注方式，能把实体边界切得很清楚，也便于模型学习不同类型实体的起始位置与跨度。  
- 2. 单层 BIO 在嵌套实体里容易“顾此失彼”。如“University of California, Los Angeles”既可整体作机构，也包含地点“Los Angeles”。同一 token 在同一层只能有一个标签，导致要么保留外层机构、丢失内层地点，要么反过来破坏整体边界。因此单层 BIO 对重叠结构表达能力有限，常需多层标注或 span 方案补充。

##### 2.2 模块 2 观察任务：1. 输入一些结构复杂的句子（包含从句或指代词，如 "He"）。观察系统（或大模型 API）是否能够准确判断出关系。2. 理解关系抽取本质上是在图结构中为两个实体节点之间预测是否存在特定语义类型的边。
- 1. 在输入含从句和指代的句子后，我发现关系抽取的稳定性明显下降。例如“Steve Jobs founded Apple, and he led it for years.”里，“Steve Jobs-Apple”的创始关系通常能识别，但“he”“it”的指代若不先做共指消解，系统很难把后续动作准确挂回原实体。句子一旦跨分句、跨代词，关系判断就不只看关键词，还依赖上下文链条。  
- 2. 关系抽取可以理解为“给知识图谱加边”的过程：NER先确定节点是谁，RE再判断节点之间是否存在某类语义连接。以“马云—阿里巴巴”为例，若预测为“FOUNDER_OF”，本质就是在人物节点与机构节点间新增一条带类型的有向边。边的类型越准，图谱检索和推理越可靠；类型错了，后续问答和分析都会被连带影响。

##### 2.3 模块 3 观察任务：输入一段包含多个人物和机构的复杂商业新闻，点击抽取后，观察生成的知识图谱是如何将“线性文本”瞬间转化为“网状结构数据”的。
- 输入一段包含多位企业家、公司及合作事件的新闻后，系统先识别出“人物”“机构”等实体，再把“创立、收购、合作、任职”等关系连成带方向的边。原本按时间顺序展开的句子，被重组为“节点-关系-节点”的网络：谁与谁直接关联、谁处于关系中心、哪些机构被多人共同连接，都能在图上立刻看清，信息结构比纯文本更直观。
