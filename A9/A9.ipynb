### 2. 实验总结

本次实验围绕“电商评论情感分析与意见挖掘平台”完成了从单条文本到批量舆情的完整链路。模块1实现了基础极性分类，并用置信度仪表盘直观展示模型判断强弱，使结果不再停留在“对或错”的二元层面。模块2进一步比较显式与隐式情感表达，验证了小模型对明显褒贬词识别更稳定，而对事实性、隐含态度语句的理解更依赖语境。模块3将方法扩展到批量评论，通过统计与可视化形成宏观口碑画像，能够直接服务于问题定位、产品优化和风险预警。整体来看，本次实验不仅加深了我对情感分析理论的理解，也让我体会到将NLP模型工程化落地时，数据组织、结果解释与业务连接同样关键。

##### 2.1 模块 1 观察任务：输入非常明显的好评和差评，观察仪表盘的变化。理解在真实工程应用中，为什么不仅需要输出分类结果，还需要输出“置信度（概率值）”。
- 在模块1测试中，输入“质量很好、物流很快”等明显好评时，仪表盘指向高分区且置信度接近1；输入“做工差、体验很糟”这类差评时，则明显转向低分区。这个变化说明模型不仅在“判别方向”，也在“表达把握程度”。工程中仅给正负标签不够，因为高置信度结果可直接触发业务动作，而低置信度结果更适合进入人工复核或延迟决策流程，从而降低误判成本。

##### 2.2 模块 2 观察任务：分别输入一句显式负面评价（如“这屏幕画质太垃圾了”）和一句隐式负面评价（如“在太阳底下根本看不清屏幕上的字”）。观察现有的小型深度学习模型能否准确“听懂”隐式表达背后的负面态度。
- 测试中，显式负面句“这屏幕画质太垃圾了”通常会被稳定识别为负面，且置信度较高；而隐式句“在太阳底下根本看不清屏幕上的字”虽然没有直接贬义词，但指向明确的使用缺陷，模型多半也能判为负面，不过置信度往往更低、波动更大。这说明小模型对显式情绪更敏感，对隐式态度的理解仍受语境建模能力限制。

##### 2.3 模块 3 观察任务：点击批量分析后，观察宏观图表的生成过程。体会将单句的情感分析技术扩展到大规模语料库时（意见挖掘），是如何为商业决策（如产品改进、危机预警）提供宏观洞察的。
- 点击批量分析后，系统先逐条完成情感判别，再把结果汇总成正、中、负三类占比，最终在饼图中呈现整体口碑结构。这个过程把“单句判断”升级为“群体画像”：企业不只知道某条评论好坏，更能看到负面是否集中、是否持续上升。基于这种宏观分布，团队可优先定位高频问题、安排产品迭代，并在负面异常波动时提前启动预警与客服干预。
