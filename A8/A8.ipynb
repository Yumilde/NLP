### 2. 实验总结

本次实验围绕“翻译机制对比与质量评测”完成了一个可交互的 Streamlit 系统，我对课堂中的机器翻译理论有了更直观的理解。通过神经机器翻译模块，我看到 NMT 能在多数场景下结合上下文完成较自然的意译；而在规则直译模块中，逐词替换虽然实现简单、可解释性强，但在长句结构重组和一词多义判断上明显受限。两者并排对比后，模型能力差异非常清晰。最后在 BLEU 评测模块中，我通过控制参考译文的语序与同义替换，观察到分数对 n-gram 匹配的敏感性，也认识到 BLEU 在“可量化比较”上的价值与“语义覆盖不足”的局限。整体而言，本次实验把“翻译生成”与“自动评价”串成了完整流程，帮助我从工程实现层面真正理解了机器翻译系统的核心问题。

##### 2.1 模块 1 观察任务：输入包含英语俚语或复杂长句的文本（如 "It rains cats and dogs."）。观察神经机器翻译能否正确处理多义词和上下文语境。
- 在观察任务中，我输入了 “It rains cats and dogs.” 及多条包含从句的长句。结果显示，神经机器翻译通常不会按字面直译“猫和狗”，而能结合上下文给出“下大雨”等更自然表达；对复杂长句也能保持主干语义完整。但当句子嵌套层次过深或含罕见俚语时，仍会出现细节偏差，说明其语境建模能力较强，但并非完全稳定。

##### 2.2 模块 2 观察任务：对比两种翻译结果。结合课件思考：基于规则的翻译在处理语序倒装（如定语从句）、一词多义时会遇到怎样不可逾越的鸿沟？
- 对比可见，基于规则的逐词替换只能完成“词面映射”，难以重建句法关系。遇到定语从句或倒装结构时，它往往按原顺序生硬拼接，导致语义主次错位；面对一词多义，也缺乏上下文判别能力，常选错义项。这种方法的瓶颈在于没有真正的语法建模与语境理解，因此很难跨越“可读”到“达意”的门槛。

##### 2.3 模块 3 观察任务：针对同一句英文，分别故意提供一个“词汇完全一样但语序全错”的中文参考，和一个“语序正确但同义词替换”的中文参考。观察 BLEU 分数的变化，体会 BLEU 算法（基于 n-gram 匹配）在评测翻译质量时的优缺点。
- 实验中我发现，若参考译文与候选译文词汇几乎一致但语序被打乱，BLEU 会因高阶 n-gram 断裂而明显下降；而语序基本正确、仅做同义词替换时，分数也会下滑，因为字面重合减少。由此可见，BLEU优点是可量化、实现简洁，对局部结构敏感；不足是过度依赖表层匹配，难充分反映语义等价与表达自然度。
